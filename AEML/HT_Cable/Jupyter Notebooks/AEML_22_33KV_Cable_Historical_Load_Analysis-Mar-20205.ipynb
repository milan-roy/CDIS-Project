```python
import pandas as pd
import os
```

```python
root_1 = os.path.join("..","Data from company")
root_2 = os.path.join("..","Data from me")
```

```python
logsheet_df = pd.read_excel(os.path.join(root_1,"Updated Logsheet Report_3 months.xlsx"),sheet_name="22_33KV")
logsheet_df = logsheet_df[['DATE','VOLTAGE_LEVEL','TO_SWITCH']]
logsheet_df
```

```python
scada_df = pd.read_csv(os.path.join(root_2,'Scada Data','Combined_Scada_Data.csv'))
scada_df = scada_df[scada_df['VOLTAGE']!='11KV'].reset_index(drop=True)
scada_df = scada_df[scada_df['PARA']=='I'].reset_index(drop=True)
scada_df
```

```python
faulty_scada_df = scada_df[scada_df['SWNO'].isin(logsheet_df['TO_SWITCH'].unique())]
faulty_scada_df
```

```python
import plotly.express as px
for swno in logsheet_df['TO_SWITCH'].unique():
    df = faulty_scada_df[faulty_scada_df['SWNO']==swno]
    fig = px.line(x = df['SYSTIME'], y=df['VALUE'])
    fig.show()
    break
```
