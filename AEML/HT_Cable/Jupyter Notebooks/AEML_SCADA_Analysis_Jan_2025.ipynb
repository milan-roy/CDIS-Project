# Combining the Scada files and Aggregatning them

```python
import os
import pandas as pd

root_dir = r'/home/milanroy/Desktop/AEML/HT_Cable/Data (from AEML)'
df = pd.DataFrame()
for file in os.listdir(os.path.join(root_dir,'SCADA')):
    temp = pd.read_csv(os.path.join(root_dir,'SCADA',file))
    temp = temp.drop(columns=['Unnamed: 0','RSNAME','EXECUTION_ID','IUUC_OPERATION','RAW_CREATED_TS','IS_DUPLICATE'])
    df = pd.concat([df,temp])

df = df.reset_index(drop=True)
df
```

```python
temp = df
```

```python
df = temp
```

```python
df = df[df['PARA'] !='Temp'].reset_index(drop=True)
df
```

```python
df['SWNO'] = df['SWNO'].astype(int)
df
```

```python
df['SYSTIME'] = pd.DataFrame(df['SYSTIME'])
df = df.sort_values(by='SYSTIME')
df
```

```python
df.to_csv('Combined_Scada_Data.csv', index=False)
```

```python
df[df['SWNO'] == 1013]
```

```python
df2 = df.groupby(['SWNO', 'PARA','VOLTAGE'])['VALUE'].mean().reset_index().rename(columns={'VALUE':'MEAN_VALUE'})
df2['MIN_VALUE'] = df.groupby(['SWNO', 'PARA','VOLTAGE'])['VALUE'].min().reset_index()['VALUE']
df2['MAX_VALUE'] = df.groupby(['SWNO', 'PARA','VOLTAGE'])['VALUE'].max().reset_index()['VALUE']
df2['STD_DEVIATION'] = df.groupby(['SWNO', 'PARA','VOLTAGE'])['VALUE'].std().reset_index()['VALUE']
df2
```

```python
## Checking if each SWNO is only associated with a specific voltage.
unique_voltages = df2.groupby('SWNO')['VOLTAGE'].nunique()
print(unique_voltages.value_counts())
```

```python
df2.to_excel('Scada_Analysis.xlsx', index=False)
df2[df2['PARA']=='I'].to_excel('Scada_Analysis-I.xlsx', index=False)
df2[df2['PARA']=='PF'].to_excel('Scada_Analysis-PF.xlsx', index=False)
df2[df2['PARA']=='MVA'].to_excel('Scada_Analysis-MVA.xlsx', index=False)
df2[df2['PARA']=='KW'].to_excel('Scada_Analysis-KW.xlsx', index=False)
```
