This notebook works on **'Fault Report 3 month'** excel sheet. There are multiple rows in the sheet with the values repeated. The code block below, read the excel sheet and keeps only the first row with a unique value in the 'EVENT_TIME' column.




```python
import pandas as pd

fault_df = pd.read_excel('/content/drive/MyDrive/CDIS Project/AEML/HT-Cable-FP/Fault Report 3 month.xlsx')
fault_df = fault_df.drop_duplicates(subset='EVENT_TIME', keep='first')
fault_df
```

## Cable Fault Nature

**Unique Cable Fault Nature and their counts and percentages.**

```python
df = pd.DataFrame(fault_df['CABLE FAULT NATURE'].value_counts())
df['Percentage'] = df['count']/df['count'].sum()*100
df

```

**Three main categories of HT Cable Faults.** \
Cable Damage: Cable Damaged  
Cable Failure: Running Cable Fault + Cable Burnt + Fault Bypassed  
Joint Failure: Casting Joint Faulty + Straight Joint Faulty + Transition Joint Faulty

```python
new_df = pd.DataFrame(df.loc['Cable Damaged',:])
new_df['Cable Failure'] = df.loc['Running Cable Fault',:] + df.loc['Cable Burnt',:] + df.loc['Fault Bypassed',:]
new_df['Joint Failure'] = df.loc['Casting Joint Faulty',:] + df.loc['Straight Joint Faulty',:] + df.loc['Transition Joint Faulty',:]
new_df = new_df.T

print("Total percentage from the main categories of cables faults: ", new_df['Percentage'].sum())
new_df
```
